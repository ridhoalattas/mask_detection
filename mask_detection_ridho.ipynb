# Inference Model

```python
import cv2
import glob
import numpy as np
import matplotlib.pyplot as plt

from ultralytics import YOLO
from ultralytics.utils.plotting import Annotator

THRESHOLD_IOU = 0.5
THRESHOLD_CONF = 0.5
MODEL_PATH = "./model/best.pt"

model = YOLO(MODEL_PATH)
    
def mask_detection(image:np.ndarray, confidence=THRESHOLD_CONF, iou=THRESHOLD_IOU) :
    """YOLOV8 inference Segmentation
    
    Params:
    - image:(np.array) 
        Initial image value.
    - confidence:(float)
        Initial confidence filter from YOLOV8 output.
    - iou:(float)
        Initial iou threshold from YOLOV8 output to remove overlapping.
    
    Returns:
    - bbox:(list)
        bounding box value.
    - conf:(list)
        bounding confidence value.
    - idx_cls:(list)
        bounding class value.
    """
    results = model(image, conf=confidence, iou=iou, verbose=False)[0]

    bbox = results.boxes.xyxy.cpu().tolist()
    conf = results.boxes.conf.cpu().tolist()
    idx_cls =  results.boxes.cls.cpu().numpy()
    idx_cls = list(int(i) for i in idx_cls)

    return bbox, conf, idx_cls
```

```python
for image_path in glob.glob("./sample/*"):
   image = cv2.imread(image_path)
   image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

   boxes, conf, idx_cls = mask_detection(image)
   annotator = Annotator(image)

   for box, cls in zip(boxes, idx_cls):
      if cls == 0 : 
         color = (255, 0, 0)
      elif cls == 1 : 
         color = (0, 255, 0)
      else : 
         color = (0, 0, 255)

      annotator.box_label(box, model.names[int(cls)], color=color)
         
   img = annotator.result() 

   plt.imshow(image)
   plt.figure(figsize=(15, 15))
   plt.show()
```
